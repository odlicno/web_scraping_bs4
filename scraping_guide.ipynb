```python
import requests
```

```python
url = "https://www.kdnuggets.com"
```

```python
res = requests.get(url)
```

```python
html_data = res.content
```

```python
print(html_data)
```

```python
from bs4 import BeautifulSoup
```

```python
parsed_data = BeautifulSoup(html_data, "html.parser")
```

```python
parsed_data.prettify()
```

```python
parsed_data.title
```

```python
parsed_data.title.string
```

```python
h2 = parsed_data.find('h2')
```

```python
h2
```

```python
h2s = parsed_data.find_all("h2")
for h2 in h2s:
    print(h2)
```

```python
h2 = parsed_data.find("h2").text
print(h2)
```

```python
h2s = parsed_data.find_all("h2")
for h2 in h2s:
    print(h2.text)
```

```python
anchors = parsed_data.find_all('a')
for a in anchors:
    print(a['href'])
```

```python
divs = parsed_data.find_all('div', class_='')
for div in divs:
    print(div)
```

```python
tags = parsed_data.find_all("li", class_="li-has-thumb")
for tag in tags:
    print(tag.text)
```

```python
data = parsed_data.select('img[alt*=KDnuggets]')
print(data)
```
